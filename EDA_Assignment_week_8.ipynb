```python
#analysis 1 Total sales by category
```

```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=category_sales, x='Category', y='Sales', palette='Blues_d')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()

print(category_sales)
```

technology has the highest number of sales

```python
#Analysis 2 - Sales by region
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=region_sales, x='Region', y='Sales', palette='Greens_d')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()

print(region_sales)
```

Western Region has the highest number of sales

```python
# Analysis 3 - Sales by Customer Segment
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(7,7))
plt.pie(segment_sales, labels=segment_sales.index, autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ff9999'])
plt.title('Sales by Customer Segment')
plt.tight_layout()
plt.show()

print(segment_sales)
```

Consumer contrubuted the largest sales.

```python
# Analysis 4 - Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=top_products, x='Sales', y='Product Name', palette='Oranges_d')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()

print(top_products)
```

```python
# Analysis 5 - Monthly Sales Trend
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Month_Year'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month_Year')['Sales'].sum().reset_index()
monthly_sales['Month_Year'] = monthly_sales['Month_Year'].astype(str)

plt.figure(figsize=(14,5))
plt.plot(monthly_sales['Month_Year'], monthly_sales['Sales'], color='blue', linewidth=1.5)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

This is called seasonality pattern. Sales seems to increase overtime with its peak around the end of the year and beginning of the year.

```python
# Analysis 6 - Sales by Ship Mode
shipmode_sales = df.groupby('Ship Mode')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=shipmode_sales, x='Ship Mode', y='Sales', palette='Purples_d')
plt.title('Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()

print(shipmode_sales)
```

standard class has the highest sales because it is the cheapest option resulting in preference of customers.

```python
# Analysis 7 - Sales Distribution
plt.figure(figsize=(10,5))
plt.hist(df['Sales'], bins=50, color='steelblue', edgecolor='white')
plt.title('Distribution of Sales')
plt.xlabel('Sales Amount ($)')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()

print("Average order value:", round(df['Sales'].mean(), 2))
print("Highest order value:", round(df['Sales'].max(), 2))
print("Lowest order value:", round(df['Sales'].min(), 2))
```

it was observed that most orders were clustered at the left side because most orders are small amounts.
